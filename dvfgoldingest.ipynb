```python

#reading data from silver 
silvermutation = spark.read.format("delta").load("/mnt/dvfsilver/mutation_silver")
display(silvermutation.limit(10))
```

```python
#create dimensions time 
dim_date=(silvermutation
.select("datemut","anneemut","moismut","jour").distinct()
.withColumnRenamed("datemut","date_id")

)
display(dim_date.limit(10))
# 4. Sauvegarder dans Gold
dim_date.write.format("delta").mode("overwrite").saveAsTable("dvfgold.dim_date")
```

```python
from pyspark.sql.functions import col, concat_ws, lpad, sha2
from pyspark.sql import DataFrame
#on pouvait utiliser la gemeotrie auusis et utiliser un windowsspec pour avoir les id

# 2. Slectionner les colonnes pour la dimension bien
dim_bien = silvermutation.select(
    col("idmutation").alias("id_bien"),
    "periodecst","libtypbien",
    "nblocmai", "nblocapt", "nblocdep", "nblocact", "nbvolmut", "nblocmut",
    "nbapt1pp", "nbapt2pp", "nbapt3pp", "nbapt4pp", "nbapt5pp",
    "nbmai1pp", "nbmai2pp", "nbmai3pp", "nbmai4pp", "nbmai5pp"
).dropDuplicates(["id_bien"])  # assure un seul bien par mutation

# 3. Affichage
display(dim_bien.limit(10))
dim_bien.write.format("delta").mode("overwrite").saveAsTable("dvfgold.dim_bien")
```

```python
from pyspark.sql.functions import row_number
from pyspark.sql.window import Window

# tape 1 : Extraire et ddupliquer les localisations
dim_localisation = (
    silvermutation
    .select("l_codinsee", "geompar", "geomparmut", "geomlocmut")
    .dropDuplicates()
)

# tape 2 : Crer un identifiant unique bas sur les 4 colonnes
window_spec = Window.orderBy(
    "l_codinsee", "geompar", "geomparmut", "geomlocmut"
)

dim_localisation = dim_localisation.withColumn(
    "id_localisation",
    row_number().over(window_spec)
)

# Rorganiser les colonnes
dim_localisation = dim_localisation.select(
    "id_localisation", "l_codinsee", "geompar", "geomparmut", "geomlocmut"
)

# Affichage
display(dim_localisation.limit(10))
dim_localisation.write.format("delta").mode("overwrite").saveAsTable("dvfgold.dim_localisation")
```

```python
dimnatmut=(silvermutation.select("libnatmut","idnatmut").distinct()
           
                                 
                                 
                                 )
display(dimnatmut)
dimnatmut.write.format("delta").mode("overwrite").saveAsTable("dvfgold.dimnatmut")
```

```python
# Jointure avec la dimension localisation
fact_mutation = silvermutation.join(
    dim_localisation.select("id_localisation", "l_codinsee", "geompar", "geomparmut", "geomlocmut"),
    on=["l_codinsee", "geompar", "geomparmut", "geomlocmut"],
    how="left"
)

# Jointure avec la dimension nature de mutation (via idnatmut ou libnatmut selon ton modle)
fact_mutation = fact_mutation.join(
    dimnatmut.select("libnatmut", "idnatmut"),
    on="idnatmut",  # ou "idnatmut" si tu prfres
    how="left"
)

fact_mutation = fact_mutation.select(
    "idmutation",               #  Cl primaire de la table de faits (mutation unique)
    "valeurfonc",               #  Mesure principale : valeur foncire de la mutation
    "sterr",                    #  Surface de terrain associe  la mutation (autre mesure)
    
    col("idmutation").alias("id_bien"),  #  FK vers la dimension bien (mme que idmutation ici car 1 bien par mutation)
    
    "id_localisation",          #  FK vers la dimension gographique (localisation des parcelles)
    "idnatmut",                #  FK vers la dimension nature de mutation (type de transaction)
    
    col("datemut").alias("id_temps")    #  FK vers la dimension temps (base ici simplement sur l'anne)
)
fact_mutation.write.format("delta").mode("overwrite").saveAsTable("dvfgold.fact_mutation")
```
