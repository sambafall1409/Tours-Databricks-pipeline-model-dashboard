```python
# üîÅ Lire les donn√©es du layer Bronze (fichier CSV)
df = spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .option("multiline", True) \
    .option("quote", '"') \
    .option("escape", '"') \
    .csv("/mnt/dvfbronze/mutation.csv")
```

```python

#je lestcoke sous format deltatable lmode overwrite ecrase l'ancien fichier par contre 
df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("dvfbronze.mutation")
```
